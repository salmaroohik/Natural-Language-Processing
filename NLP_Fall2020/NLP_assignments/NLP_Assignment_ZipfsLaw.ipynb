```python
#Question 1 (2 points): Load the course data files (both English and Italian) into dataframes. Each data file should be in a separate dataframe. 
    
    
import pandas as pd
df_en=pd.read_table('c:\\Users\\salma\\Downloads\\CONcreTEXT_trial_EN.tsv') #English dataframe 
df_it=pd.read_table('c:\\Users\\salma\\Downloads\\CONcreTEXT_trial_IT.tsv') #Italian dataframe 
```

```python

#Question 2 (2 points): Using NLTK libraries, strip the text of punctuation and lowercase the words, and obtain from the original text a list of words.
import nltk
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokenized_list_en=[]
word_en=[]
for text in df_en["TEXT"]:
    tokenized_list_en=tokenizer.tokenize(text.lower())
    for word in tokenized_list_en:
        word_en.append(word)
print("English words:" )
print('\n')
print(word_en)

print('\n')
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokenized_list_it=[]
word_it=[]
for text in df_it["TEXT"]:
    tokenized_list_it=tokenizer.tokenize(text.lower())
    for word in tokenized_list_it:
        word_it.append(word)


print("Italian  words:" )
print('\n')
print(word_it)
    

```

Question 6 (5 points): Discuss whether the text satisfies Zipf‚Äôs law or not. Why do you think that is the case? Please provide an *original* answer with least 100 words. Answer: The text here satisfies Zipf‚Äôs law.
Zipf's formulated in terms of quantitative linguistics states, given some corpus of natural language utterances, 

The frequency of any word is inversely proportional to its rank in the frequency table. 
Thus the most frequent word will occur approximately twice as often as the second most frequent word, three times as often as the third most frequent word, etc.: the rank-frequency distribution is an inverse relation

Here in the given english dataset, 'the': {'rank': 1, 'occurrences': 63}, 'you': {'rank': 2, 'occurrences': 45}, 'a': {'rank': 3, 'occurrences': 39} are the top 3 words with rank and frequency tabulated respectively. The occurrence of "the" is approximately twice nd and thirce the latter words. The same relationship is seen in graph "plot English (a)" and "plot Italian (a)". The rank-frequency distribution inverse corelation is satisfied throughout the plot. 

```python
#***Question 7 (4 points): Plot (a) frequency vs. rank and (b) log frequency vs. log rank in order to "visualize" Zipf's Law for each corpus (English and Italian). Note: This will be four plots total.
list_plot=[]
%matplotlib inline
from matplotlib import pyplot as plt
for x in final_en:
    df=final_en[x]['occurrences'],final_en[x]['rank']
    list_plot.append(df)
#print(list_plot)

import pandas as pd 
data = final_en 
df = pd.DataFrame(list_plot) 


#pd.DataFrame(df).plot()
plt.plot(df[1],df[0], marker=True)
plt.xlabel('Rank ')
plt.ylabel('frequency')
plt.title('Plot English (a) frequency vs. rank ')
plt.show()

```

```python
import math
list_log_f=[]
list_log_r=[]
for i in df[0]:
    x=math.log10(i)
    list_log_f.append (x)
    
for j in df[1]:
    y=math.log10(j)
    list_log_r.append (y)
    
df[2]=list_log_f
df[3]=list_log_r

plt.plot(df[3],df[2])
plt.xlabel('Log of Rank ')
plt.ylabel('Log of frequency')
plt.title('Plot English (b) log frequency vs. log rank  ')
plt.show()
```
